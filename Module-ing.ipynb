```python
ohe = pd.get_dummies(ch, sparse=False)
ohe
```

```python
ohe["제휴사_A_MOBILE/APP"] = ohe['이용횟수'] * ohe["제휴사_A_MOBILE/APP"]
ohe["제휴사_B_MOBILE/APP"] = ohe['이용횟수'] * ohe["제휴사_B_MOBILE/APP"]
ohe["제휴사_C_MOBILE/APP"] = ohe['이용횟수'] * ohe["제휴사_C_MOBILE/APP"]
ohe["제휴사_D_MOBILE/APP"] = ohe['이용횟수'] * ohe["제휴사_D_MOBILE/APP"]
ohe["제휴사_B_ONLINEMALL"] = ohe['이용횟수'] * ohe["제휴사_B_ONLINEMALL"]
ohe["제휴사_C_ONLINEMALL"] = ohe['이용횟수'] * ohe["제휴사_C_ONLINEMALL"]
ohe
```

```python
ohe.drop('이용횟수', axis=1, inplace=True)
```
